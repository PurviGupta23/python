```python
l = [10, 20, 30, 40, 50]
print(l)
# Indexing - Accessing elements of List using index number 
# Syntax: list_name[index]
print(l[0]) # 10
print(l[1]) # 20
print(l[2]) # 30
print(l[3]) # 40
print(l[4]) # 50
# print(l[5]) # IndexError: list index out of range
print(l[-1]) # 50
print(l[-2]) # 40
print(l[-3]) # 30
print(l[-4]) # 20
print(l[-5]) # 10
# print(l[-6]) # IndexError: list index out of range
```

```python
# Slice - Accessing range of elements from List using slice operator : 
# Syntax: list_name[start_index : end_index] or [inclusive, exclusive]
# start_index is inclusive and end_index is exclusive
# eg: if I have to print 20, 30, 40 then start_index = 1 and end_index = 4
print(l[1:4]) # [20, 30, 40]
print(l[0:2]) # [10, 20]
print(l[2:5]) # [30, 40, 50]
print(l[0:5]) # [10, 20, 30, 40, 50]
print(l[0:6]) # [10, 20, 30, 40, 50] - end_index is greater than length of list, it will print all elements
print(l[0:]) # [10, 20, 30, 40, 50] - end_index is not provided, it will print all elements
print(l[3:]) # [40, 50] - start_index is not provided, it will print all elements from start_index
print(l[:3]) # [10, 20, 30] - if start_index is not provided, it will start from 0th index and print till end_index-1
```

```python
# Negative Slice - Accessing range of elements from List using slice operator :
# Syntax: list_name[start_index : end_index]
# start_index is inclusive and end_index is exclusive
# eg: if I have to print 20, 30, 40 then start_index = -4 and end_index = -1
print(l[-4:-1]) # [20, 30, 40]
print(l[-1:-4] ) # [] - start_index is greater than end_index, it will return empty list
print(l[-1:5]) # [50] - end_index is greater than length of list, it will print all elements from start_index
print(l[-3:4]) # [30, 40] - start_index is negative and end_index is positive

# print last 5 elements of list
# Syntax: list_name[start_index : end_index : step]
# step is the number of elements to skip while slicing. By default, it is 1 
print(l[-1:-4:-1]) # [50, 40, 30]
print(l[::-1]) # [50, 40, 30, 20, 10] - reverse the list
print(l[ : : -2 ]) # [50, 30, 10] - reverse the list and skip 1 element - Syntax: list_name[start_index : end_index : step]
```

```python
# Syntax: list_name.function_name()
l = [10, 20, 30, 40, 50]
print(l) # [10, 20, 30, 40, 50]
```

```python
# append() - Add an element at the end of the List
# Syntax: list_name.append(element)
l.append(60) # [10, 20, 30, 40, 50, 60]
print(l)
```

```python
# insert() - Add an element at a specific index
# Syntax: list_name.insert(index, element)
l.insert(2, 25) # [10, 20, 25, 30, 40, 50, 60]
print(l)
```

```python
# remove() - Remove an element from the List
# Syntax: list_name.remove(element)
l.remove(25) # [10, 20, 30, 40, 50, 60]
print(l)
```

```python
# pop() - Remove an element from the List at a specific index
# Syntax: list_name.pop(index)
l.pop(3) # [10, 20, 30, 50, 60]
print(l)
```

```python
# clear() - Remove all elements from the List
# Syntax: list_name.clear()
l.clear() # []
print(l)
```

```python
# del - Remove the List
# Syntax: del list_name
# del is a keyword in Python which is used to delete the reference of the object from the memory
l = [10, 20, 30, 40, 50]
print(l)     # [10, 20, 30, 40, 50]
del l      # print(l) # NameError: name 'l' is not defined
```

```python
# extend() - Add elements of one List to another List
# Syntax: list_name.extend(list_name)
l1 = [10, 20, 30]
l2 = [40, 50, 60]
l1.extend(l2)     # [10, 20, 30, 40, 50, 60]
print(l1)         # [10, 20, 30, 40, 50, 60]
print(l2)         # [40, 50, 60]
```

```python
# Sort() - Sort the elements of List
# Syntax: list_name.sort()
l = [40, 10, 30, 20, 50]
print(l)         # [40, 10, 30, 20, 50]
l.sort()         # [10, 20, 30, 40, 50]
print(l)
```

```python
# Sort() - Sort the elements of List in descending order
# Syntax: list_name.sort(reverse=True)
# reverse is by default False. If you want to sort the elements in descending order, then set reverse=True
l = [40, 10, 30, 20, 50]
print(l)               # [40, 10, 30, 20, 50]
l.sort(reverse=True)   # [50, 40, 30, 20, 10]
print(l)
```

```python
# sort() vs sorted()
# sort() - sort the elements of List in ascending order. It changes the original List and returns None
# sort() is a function of List data structure
# sorted() - sort the elements of List in ascending order. It does not change the original List and returns a new List
# sorted() is a built-in function in Python, and is not a function of List data structure. It is a general purpose function
l = [40, 10, 30, 20, 50]
print(l)           # [40, 10, 30, 20, 50]
l1 = sorted(l)
print(l)           # [40, 10, 30, 20, 50]
print(l1)          # [10, 20, 30, 40, 50]
```

```python
l = [40, 10, 30, 20, 50]
print(l)        # [40, 10, 30, 20, 50]
l.sort()
print(l)        # [10, 20, 30, 40, 50]
```

## Set { }

- Set - It is a collection which is unordered and unindexed. No duplicate members.
- In Set, items are stored in curly brackets { } and separated by commas.
- Set can have any number of items and they may be of different types (integer, float, string etc.)
- Set is used to store multiple items in a single variable.

- .add(element) - Add an element to the Set
- .update(element) - Add multiple elements to the Set
- .update([element1, element2, element3....])
- .remove(element) - Remove an element from the Set
- .discard(element) - Remove an element from the Set
- .pop() - Remove an arbitary element from the Set
- .clear() - Remove all elements from the Set
- new_set = set1.intersection(set2) - Returns a set, that is the intersection of two other sets
- new_set = set1.difference(set2) - Returns a set containing the difference between two or more sets
- subtract() - Returns a set containing the difference between two or more sets
- list_name = list(set) - Convert set to list
- print(A | B)   # Union              print(A.union(B))
- print(A & B)   # Intersection       print(A.intersection(B))
- print(A - B)   # Difference         print(A.difference(B))     # only A elements
- print(A ^ B)   # Symmetric Difference
- new_set = frozenset(old_set)      Frozen set
- list_name = list(set_name)    -  Convert set to list
- https://www.w3schools.com/python/python_ref_set.asp

```python
# Declare a Set
s = {219, "Darshan", 3.1416, True, -87, 321}     # Set of mixed data types
print(s)                                         # Set is unordered i.e. the items will appear in a random order
print(type(s))
# print(s[0])                                    # TypeError: 'set' object is not subscriptable because Set is unindexed
```

```python
# No duplicate members are allowed in Set
s = {10, 20, 30, 40, 50, 10, 20, 30, 40, 50, 10, 20, 30, 40, 50}
print(s)                                           # {40, 10, 50, 20, 30} - No duplicate members are allowed in Set
```

```python
# Functions which can be applied on Set
# Syntax: set_name.function_name()
s = {10, 20, 30, 40, 50}
print(s)                      # {40, 10, 50, 20, 30}
```

```python
# add() - Add an element to the Set
# Syntax: set_name.add(element)
s.add(60)                # {40, 10, 50, 20, 60, 30}
print(s)
```

```python
# update() - Add multiple elements to the Set
# Syntax: set_name.update([element1, element2, element3, ...])
s.update([70, 80, 90])                        # {70, 40, 10, 80, 50, 20, 90, 60, 30}
print(s)
```

```python
# remove() - Remove an element from the Set
# Syntax: set_name.remove(element)
s.remove(70)                                  # {40, 10, 80, 50, 20, 90, 60, 30}
print(s)
```

```python
# discard() - Remove an element from the Set
# Syntax: set_name.discard(element)
s.discard(80)                               # {40, 10, 50, 20, 90, 60, 30}
print(s)
```

```python
# pop() - Remove an element from the Set
# Syntax: set_name.pop()
s.pop()                                  # {10, 50, 20, 90, 60, 30}  - pop() removes the arbitary element from the Set
print(s)
s.pop()                                   # {50, 20, 90, 60, 30}
print(s)
```

```python
# clear() - Remove all elements from the Set
# Syntax: set_name.clear()
s.clear() # set()
print(s)
```

```python
# intersection() - Returns a set, that is the intersection of two other sets
X = {"Apple", "Banana", "Cherry"}
Y = {"Google", "Microsoft", "Apple"}
Z = X.intersection(Y)
print(Z)                           # {'Apple'}
```

```python
# subtract() - Returns a set containing the difference between two or more sets
X = {"Google", "Microsoft", "Apple"}
Y = {"Apple", "Banana", "Cherry"}
Z = X.difference(Y)
print(Z)                              # {'Google', 'Microsoft'}
```

```python
# Convert set to list
X = {"Google", "Microsoft", "Apple"}
print(X)                  # {'Google', 'Microsoft', 'Apple'}
l = list(X)
print(l)                      # ['Google', 'Microsoft', 'Apple']
```

## Dictionary { }

- In Python, we use dictionaries to store key-value pairs. Here is an example - Let's take the example of a customer as seen earlier. Now a customer could have multiple attributes such as

- Name
- Email
- Userid
- Address
- Phone number
- Now each of these attributes would be our key and corresponding we'll have values against each on them

- With a dictionary, we can store a bunch of key-value pairs, let us now see have a look at the 'Dictionary' data type in Python

Dictionaries 
- A dictionary stores element as keys and values pairs.
- The key is like an index, it's is always unique and immutable.
- The values are the objects that contain information.
- Values are accessed using their keys.
- Each key is followed by a value separated by a colon.
- The values can be immutable, mutable, and duplicates.

- 
-  Dictionary - It is a collection which is unordered, changeable and indexed. No duplicate members.
-  Syntax: dict_name = {key1: value1,   key2: value2,    key3: value3,   ...}
-  In Dictionary, items are stored in curly brackets { } and separated by commas.
-  Dictionary has keys and values. Keys are unique and values can be duplicate.
-  Dictionary can have any number of items and they may be of different types (integer, float, string etc.)

- dict_name["key_name"]                       - to Access value using keys
- dict_name["key_name"] = "new_value"         - Replace the value for a key in a dictionary
- dict_name["new_key_name"] = "new_value"     - Insert new key value pair into a dictionary
- del dict_name["key_name"]                    - Deleting a key value pair
- sorted(dict_name)                            - Sorting a dictionary
- dict_name.values()                           - to access all the values
- dict_name.keys()                             - to access all the keys
- dict_name.update({key:value})                - to update key value pair
- dict_name.items()                             - all keys and values will be output in list of tuples

- get()
- clear()
- copy()
- items()
- keys()
- .pop(key_name)  pop need the key to remove the value and key, only one key at a time
- divt_name['new_name'] = dict_name.pop('old_key_name')       to rename any key or changing the key name with the old values
- values()
- .popitems()    -  always pop out last values by default
- .setdefault('new_key_name')    -  want a new key but don't know about the values right now , later we can update that
- stu_details
- update()

```python
# Functions which can be applied on Dictionary
# Syntax: dict_name.function_name()
l = {
    0 : 10,
    1 : 20,
    2 : 30,
    3 : 40,
    4 : 50
}
print(l) # {0: 10, 1: 20, 2: 30, 3: 40, 4: 50}
```

```python
# get() - Returns the value of the specified key
# Syntax: dict_name.get(key)
print(l.get(0)) # 10
print(l.get(4)) # 50
```

```python
# items() - Returns a list containing a tuple for each key value pair
# Syntax: dict_name.items()
print(l.items()) # dict_items([(0, 10), (1, 20), (2, 30), (3, 40), (4, 50)])
print(list(l.items())) # [(0, 10), (1, 20), (2, 30), (3, 40), (4, 50)]
```

```python
# keys() - Returns a list containing the dictionary's keys
# Syntax: dict_name.keys()
print(l.keys()) # dict_keys([0, 1, 2, 3, 4])
print(list(l.keys())) # [0, 1, 2, 3, 4]
```

```python
# values() - Returns a list containing the dictionary's values
# Syntax: dict_name.values()
print(l.values()) # dict_values([10, 20, 30, 40, 50])
print(list(l.values())) # [10, 20, 30, 40, 50]
```
